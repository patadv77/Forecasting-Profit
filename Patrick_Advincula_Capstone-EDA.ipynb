As we have missing values here we can implore imputing techniques

This is a great explaintion for the Timeseries notebook for class

Next, to impute the missing values we have three common options:
- `forward filling`: the missing value is filled by the last valid column value,
- `backward filling`: the missing value is filled by the next following valid observation,
- `interpolation`: use previous and following value(s) to interpolate the missing one (e.g., using the mean or some linear combination like a rolling average).

# Poisson regression on my categorical variables as my Target value is quantitative discrete

```python
from scipy.stats import chi2_contingency
```

```python
import statsmodels.api as sm
```

```python

# Poisson

# Specify the categorical predictors and the outcome variable
predictors = ['Pricing::Segment',
 'Promotion::SKU',
 'SOP_Half',
 'SOP_Quarter']
outcome = 'Final Quantity'

# Create dummy variables for the categorical predictors
dummy_skudata_interpolated = pd.get_dummies(skudata_interpolated[predictors], drop_first=True)

# Add the outcome variable to the dummy skudata_interpolated
dummy_skudata_interpolated[outcome] = skudata_interpolated[outcome]

# Fit the Poisson regression model
model = sm.GLM(dummy_skudata_interpolated[outcome], dummy_skudata_interpolated[dummy_skudata_interpolated.columns[:-1]], family=sm.families.Poisson()).fit()

# Print the summary of the model
print(model.summary())
```

# Now lets test do a statistical test on our quantiative discrete values. Here I will use a OLS regression.

```python
import pandas as pd
import statsmodels.api as sm
```

```python
skudata_interpolated.select_dtypes(exclude=object).columns.to_list()
```

```python
# Specify the predictor variables and the outcome variable
X = skudata_interpolated[['Pricing::Street',
 'Promotion::Promo_Value',
 'Promotion::Promo Type',
 'Promotion::Promo_Net Price Point',
 'Account',
 'Sell Through Estimate']]
y = skudata_interpolated['Final Quantity']
```

```python

# Add a constant term to the predictor variables
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression analysis
print(model.summary())
```

# To visiualize the "Final Qty" Histogram

```python
import matplotlib.pyplot as plt

# Group by 'Promotion::SKU' & 'Final Quantity'
grouped = skudata_interpolated.groupby('Promotion::SKU')

# Loop through each group and create a histogram
for sku, group in grouped:
    # Compute mean and median
    mean = group['Final Quantity'].mean()
    median = group['Final Quantity'].median()
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(20, 12))
    ax.hist(group['Final Quantity'], bins=100)
    
    # Add title and labels
    ax.set_title(sku)
    ax.set_xlabel('Final Quantity')
    ax.set_ylabel('Frequency')
    
    # Add mean and median lines
    ax.axvline(x=mean, color='r', linestyle='--', label=f'Mean: {mean:.2f}')
    ax.axvline(x=median, color='b', linestyle='--', label=f'Median: {median:.2f}')
    
    ax.legend()
    plt.show()
```
